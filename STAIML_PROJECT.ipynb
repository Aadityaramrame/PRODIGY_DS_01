```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def linear_regression(x, y):
    n = len(x)
    # Calculate the sum of x, y, x^2, xy
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_x_squared = np.sum(x**2)
    sum_xy = np.sum(x * y)
    # Calculate the coefficient (slope and intercept)
    slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
    intercept = (sum_y - slope * sum_x) / n
    # Return the coefficients
    return {'intercept': intercept, 'slope': slope}

# Load dataset

data = pd.read_csv('/content/Car_analysis.csv')  # Assuming the file is in the same directory as the script

# Select columns for linear regression
x = data['mileage(kmpl)']
y = data['max_power(bhp)']

coefficients = linear_regression(x, y)
print(coefficients)

# Plot the data and the linear regression line
plt.scatter(x, y, color='red', marker='o', label='Data Points')
plt.plot(x, coefficients['intercept'] + coefficients['slope'] * x, color='blue', linewidth=2, label='Linear Regression Line')
plt.xlabel('Mileage (kmpl)')
plt.ylabel('Max Power (bhp)')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load the dataset
file_path = "/content/Car_analysis.csv"
data = pd.read_csv(file_path)

# Customer Segmentation based on Preferred Fuel Type (Frequency)
fuel_type_segmentation = data['fuel_type'].value_counts()

# Customer Segmentation based on Transmission Type (Frequency)
transmission_segmentation = data['transmission'].value_counts()

# Define price range categories
price_ranges = {
    'Low': (0, 5),   # Example: cars priced up to 5 lakhs
    'Medium': (5, 15),  # Example: cars priced between 5 and 15 lakhs
    'High': (15, float('inf'))  # Example: cars priced above 15 lakhs
}

# Function to categorize price range
def categorize_price(price):
    for range_name, (min_price, max_price) in price_ranges.items():
        if min_price <= price <= max_price:
            return range_name

# Apply the categorization function to create a new column 'Price_Range'
data['Price_Range'] = data['price(in lakhs)'].apply(categorize_price)

# Customer Segmentation based on Price Range (Frequency)
price_range_segmentation = data['Price_Range'].value_counts()

# Display the segmentation results
print("Customer Segmentation based on Preferred Fuel Type (Frequency):")
print(fuel_type_segmentation)
print("\nCustomer Segmentation based on Transmission Type (Frequency):")
print(transmission_segmentation)
print("\nCustomer Segmentation based on Price Range (Frequency):")
print(price_range_segmentation)
```
